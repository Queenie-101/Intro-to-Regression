```python
import matplotlib.pyplot as plt #from matplotlib import pyplot
import numpy as np
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error #check for errors and metrics
import pandas as pd
```

```python
#load the dataset
diabetes_X,diabetes_y=datasets.load_diabetes(return_X_y=True)
```

```python
#use only one feature
diabetes_x=diabetes_X[:,np.newaxis,2]
```

```python
#split data
diabetes__X_train=diabetes_X[:-20]
diabetes_X_test=diabetes_X[-20:]
diabetes__y_train=diabetes_y[:-20]
diabetes_y_test=diabetes_y[-20:]
```

```python
#create the linear model
model=linear_model.LinearRegression()
```

```python
#train the algorithm/model with the data
model.fit(diabetes__X_train,diabetes__y_train)
```

```python
#make predictions using the algorithm
predicted_values=model.predict(diabetes__X_test)
```

```python
#compare the predicted values and original values in a dataframe
df=pd.DataFrame({'Actual':diabetes__y_test,'Predicted':predicted_values})
```
